```python
import os
import json
import re
import requests
import warnings
from kvt_extract_ratecon_werner import extract
from kvt_extract_ratecon_werner import references_sh
from kvt_extract_ratecon_werner import references_mc
from kvt_extract_ratecon_werner import references_dot
from kvt_extract_ratecon_werner import references_bol
from kvt_extract_ratecon_werner import references_crpo
from kvt_extract_ratecon_werner import order_detail
from kvt_extract_ratecon_werner import date_pick
from kvt_extract_ratecon_werner import entities
from kvt_extract_ratecon_werner import entities_drop
from kvt_extract_ratecon_werner import entities_broker
from kvt_extract_ratecon_werner import note
from kvt_extract_ratecon_werner import stop_pick
from kvt_extract_ratecon_werner import stop_drop
from kvt_extract_ratecon_werner import date_drop2
from kvt_extract_ratecon_werner import date_drop3
from kvt_extract_ratecon_werner import date_drop4
from kvt_extract_ratecon_werner import date_drop5
from kvt_extract_ratecon_werner import date_drop6
from kvt_extract_ratecon_werner import date_drop7
```

```python
# get json from url

def get_json_from_url(url: str)-> dict:
    try:
        response = requests.get(url)
        
        if response.status_code == 200:
            return json.loads(response.text)
        else:
            warnings.warn('Fail to get json data from url, status code:{status}.'.format(status=response.status_code))
            return None
    except Exception as e:
        warnings.warn('Fail to get json data from url, exception{exception}'.format(exception=e))
        return None
```

```python
#data clean, remove spaces, seperate by ;
def data_clean(text):
    try:
        strip_po = [i.strip() for i in text.split('\n')]
        remove_space = [i for i in strip_po if i not in[' ', '']]

        cleaned_data = list()
        for s in remove_space:
            cleaned_data.append(
                    re.sub('\s+', ' ', s)
                 )
        
        return ';'.join(cleaned_data)
    except Exception as e:
        print("kvt data clean exception")
        print(e)
        return None
```

```python
def process_data(text_map):
    details = extract(text_map)
    details_references_sh = references_sh(text_map)
    details_references_dot = references_dot(text_map)
    details_references_bol = references_bol(text_map)
    details_references_mc = references_mc(text_map)
    details_references_crpo = references_crpo(text_map)
    details_date_pick = date_pick(text_map)
    details_entities = entities(text_map)
    details_entities_drop = entities_drop(text_map)
    details_entities_broker = entities_broker(text_map)
    details_note = note(text_map)
    details_stop_pick = stop_pick(text_map)
    details_stop_drop = stop_drop(text_map)
    details_stops_order_detail = order_detail(text_map)
    details_date_drop2 = date_drop2(text_map)
    details_date_drop3 = date_drop3(text_map)
    details_date_drop4 = date_drop4(text_map)
    details_date_drop5 = date_drop5(text_map)
    details_date_drop6 = date_drop6(text_map)
    details_date_drop7 = date_drop7(text_map)
    details['references'].append(details_references_sh)
    details['references'].append(details_references_dot)
    details['references'].append(details_references_bol)
    details['references'].append(details_references_mc)
    details['references'].append(details_references_crpo)
    details['dates'].append(details_date_pick)
    details['dates'].append(details_date_drop2)
    details['dates'].append(details_date_drop3)
    details['dates'].append(details_date_drop4)
    details['dates'].append(details_date_drop5)
    details['dates'].append(details_date_drop6)
    details['dates'].append(details_date_drop7)
    details['entities'].append(details_entities)
    details['entities'].append(details_entities_drop)
    details['entities'].append(details_entities_broker)
    details['notes'].append(details_note)
    details['stops'].append(details_stop_pick)
    details['stops'].append(details_stop_drop)   
    details['stops'][0]["dates"].append(details_date_pick)
    details['stops'][1]["dates"].append(details_date_drop2)
    details['stops'][1]["dates"].append(details_date_drop3)
    details['stops'][1]["dates"].append(details_date_drop4)
    details['stops'][1]["dates"].append(details_date_drop5)
    details['stops'][1]["dates"].append(details_date_drop6)
    details['stops'][1]["dates"].append(details_date_drop7)
    details['stops'][0]["notes"].append(details_note)
    details['stops'][0]["entities"].append(details_entities)
    details['stops'][0]["entities"].append(details_entities_broker)
    details['stops'][1]["entities"].append(details_entities_drop)
    details['stops'][1]["entities"].append(details_entities_broker)
    details['stops'][0]["references"].append(details_references_sh)
    details['stops'][0]["references"].append(details_references_dot)
    details['stops'][0]["references"].append(details_references_bol)
    details['stops'][0]["references"].append(details_references_mc)
    details['stops'][0]["references"].append(details_references_crpo)
    details['stops'][0]["order_detail"][0]["purchase_order_number"].append(details_stops_order_detail)
    details['stops'][1]["notes"].append(details_note)
    details['stops'][1]["references"].append(details_references_sh)
    details['stops'][1]["references"].append(details_references_dot)
    details['stops'][1]["references"].append(details_references_bol)
    details['stops'][1]["references"].append(details_references_mc)
    details['stops'][1]["references"].append(details_references_crpo)
    details['stops'][1]["order_detail"][0]["purchase_order_number"].append(details_stops_order_detail)

    


    return(details)
    
```

```python
#result 
def result():  
    with open('data/text_data/cornerstone_systems_text_data_links.json', 'r') as myfile:
        data=myfile.read()
    data_list=json.loads(data)
    
    for i in data_list:
        
        
        a = get_json_from_url(i)
        for key in a["text"]:
            b=a["text"][key]
            a["text"][key]=data_clean(b)
            kvt=process_data(a["text"])
        print(i)
        print(json.dumps(kvt, indent=4))
        

result()    
```
